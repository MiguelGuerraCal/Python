```python
# Importar las bibliotecas necesarias
import yfinance as yf
import pandas as pd

# Crear un objeto de la acción Tesla
tesla = yf.Ticker("TSLA")

# Extraer datos históricos de Tesla con el período máximo
tesla_data = tesla.history(period="max")

# Restablecer el índice del DataFrame
tesla_data.reset_index(inplace=True)

# Mostrar las cinco primeras filas
print(tesla_data.head())
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
# URL de la página web con los datos de ingresos de Tesla
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
# Enviar solicitud a la URL y analizar el contenido HTML
html_data = requests.get(url).text
soup = BeautifulSoup(html_data, "html.parser")
# Crear un DataFrame vacío para almacenar los datos
tesla_revenue = pd.DataFrame(columns=["Date", "Revenue"])
# Buscar la tabla en la página web
for row in soup.find("tbody").find_all("tr"):
    col = row.find_all("td")
    date = col[0].text
    revenue = col[1].text
    tesla_revenue = pd.concat(
        [tesla_revenue, pd.DataFrame({"Date": [date], "Revenue": [revenue]})],
        ignore_index=True,
    )
# Mostrar las últimas cinco filas
print(tesla_revenue.tail())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
fechas = [datetime.today() - timedelta(days=i*30) for i in range(12)]  # Últimos 12 meses
ingresos = [5000 + i * 100 for i in range(12)]
tesla_revenue = pd.DataFrame({'Date': fechas, 'Revenue': ingresos})
acciones_fechas = [datetime.today() - timedelta(days=i) for i in range(30)]  # Últimos 30 días
acciones_precios = [200 + i for i in range(30)]
tesla_data = pd.DataFrame({'Date': acciones_fechas, 'Close': acciones_precios})
# Definir la función para graficar
def make_graph(datos_acciones, datos_ingresos, acciones):
    fig, ax1 = plt.subplots(figsize=(12, 6))
    ax1.set_xlabel('Fecha')
    ax1.set_ylabel('Precio de Acciones (USD)', color='tab:blue')
    ax1.plot(datos_acciones['Date'], datos_acciones['Close'], label=f'Acciones de {acciones}', color='tab:blue')
    ax1.tick_params(axis='y', labelcolor='tab:blue')
    ax1.grid(True)
    ax2 = ax1.twinx()
    ax2.set_ylabel('Ingresos (USD)', color='tab:green')
    ax2.plot(datos_ingresos['Date'], datos_ingresos['Revenue'], label=f'Ingresos de {acciones}', color='tab:green')
    ax2.tick_params(axis='y', labelcolor='tab:green')
    plt.title(f'Datos de {acciones}: Acciones e Ingresos')
    fig.tight_layout()
    plt.show()
make_graph(tesla_data, tesla_revenue, 'Tesla')
```
